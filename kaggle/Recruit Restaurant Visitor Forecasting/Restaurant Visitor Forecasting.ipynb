### Creating master dataset

```python
master = air_visit_data.copy()
master.head()
```

```python
master = pd.merge(master, air_store_info, on='air_store_id')
```

```python
master.head()
```

```python
master.shape
```

```python
len(master['air_store_id'].unique())
```

```python
master.isnull().sum()
```

```python
master = pd.merge(master, date_info, on='visit_date')
```

```python
master.drop(['latitude','longitude'],axis=1, inplace=True)
```

```python
master.head()
```

```python
master.shape
```

```python
master.dtypes
```

```python
master['month'] = 0
for i, row in master.iterrows():
    vals  = row['visit_date'][5:7]
    master.at[i, 'month'] = vals
```

```python
master['year'] = 0
for i, row in master.iterrows():
    vals  = row['visit_date'][:4]
    master.at[i, 'year'] = vals
```

```python
mnth = pd.get_dummies(master['month'])
mnth.head()
```

```python
master = pd.concat([master, mnth], axis=1)
```

```python
year = pd.get_dummies(master['year'])
year.head()
```

```python
master = pd.concat([master, year], axis=1)
```

```python
dow = pd.get_dummies(master['day_of_week'])
master = pd.concat([master, dow], axis=1)
```

```python
agn = pd.get_dummies(master['air_genre_name'])
master = pd.concat([master, agn], axis=1)
```

```python
aan = pd.get_dummies(master['air_area_name'])
master = pd.concat([master, aan], axis=1)
```

```python
test_df['air_store_id'] = '0'
test_df['visit_date'] = '0'
test_df.columns
```

```python
for i, row in test_df.iterrows():
    var = row['id'].split('_')
    test_df.at[i,'air_store_id'] = var[0] +'_'+ var[1]
    test_df.at[i,'visit_date'] = var[2]    
```

```python
test_df = pd.merge(test_df, air_store_info, on='air_store_id')
```

```python
test_df = pd.merge(test_df, date_info, on='visit_date')
```
