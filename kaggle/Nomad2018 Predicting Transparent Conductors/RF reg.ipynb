```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import ParameterGrid
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer
from sklearn.preprocessing import MinMaxScaler
```

```python
train = pd.read_csv('train_v2.csv')
train.head(2)
```

```python
test = pd.read_csv('test_v2.csv')
test.head(2)
```

```python
train.id = train.id.astype(int)
```

```python
train.set_index('id',inplace=True)
test.set_index('id',inplace=True)
```

```python
full_train = train.columns.tolist()
full_test = test.columns.tolist()
full_train.remove('formation_energy_ev_natom')
full_train.remove('bandgap_energy_ev')
X = full_train
y = ['formation_energy_ev_natom','bandgap_energy_ev']
```

```python
X_data = train[X]
y_data = train[y]
```

```python
X1_train, X1_val, y1_train, y1_val = train_test_split(X_data, y_data['formation_energy_ev_natom'], test_size=0.2, shuffle=True)
```

```python
X2_train, X2_val, y2_train, y2_val = train_test_split(X_data,  y_data['bandgap_energy_ev'], test_size=0.2, shuffle=True)
```

```python
mae_score = make_scorer(mean_absolute_error, greater_is_better=False)
```

```python
rf_grid = {
        'n_estimators': [10,15,20,25,30,50,75,100],
        'min_samples_leaf': [5,10,15,20,25,30,35,40,45,50],
        'n_jobs': [-1],
        'verbose': [1]
    }
rfr = RandomForestRegressor()
```

```python
rfr_gs1 = GridSearchCV(rfr, param_grid = rf_grid, n_jobs=-1, cv=3,  scoring =  mae_score, verbose=1)
```

```python
rfr_gs1.fit(X1_train,y1_train)
```

```python
gs1 = [x[1] for x in rfr_gs1.grid_scores_]
gs1_filter = [i for i in gs1 if i > -0.027]
gs1_filter
```

```python
gs1_model = list()
for x  in rfr_gs1.grid_scores_:
    if x[1] > -0.027:
        gs1_model.append(x)
gs1_model  
```

```python
rfr_m1 = RandomForestRegressor(min_samples_leaf=5, n_estimators=100, n_jobs=-1)
rfr_1 = rfr_m1.fit(X1_train,y1_train)
```

```python
p1 = rfr_m1.predict(X1_val)
p1.shape
```

```python
print(mean_absolute_error(y1_val,p1))
```

```python
rfr_gs2 = GridSearchCV(rfr, param_grid = rf_grid, n_jobs=-1, cv=5,  scoring =  mae_score, verbose=1)
```

```python
rfr_gs2.fit(X2_train,y2_train)
```

```python
gs2 = [x[1] for x in rfr_gs2.grid_scores_]
gs2_filter = [i for i in gs2 if i > -0.16]
gs2_filter
```

```python
gs2_model = list()
for x  in rfr_gs2.grid_scores_:
    if x[1] > -0.16:
        gs2_model.append(x)
gs2_model 
```

```python
rfr_m2 = RandomForestRegressor(min_samples_leaf=5, n_estimators=100, n_jobs=-1)
rfr_2 = rfr_m2.fit(X2_train,y2_train)
p2 = rfr_m2.predict(X2_val)
p2.shape
```

```python
print(mean_absolute_error(y2_val,p2))
```

### Combining NN & RF

```python
nn = pd.read_csv('sub2_nn.csv')
nn.head()
```

```python
rf = pd.read_csv('sub1_rf.csv')
rf.head()
```

```python
fl = ['id','formation_energy_ev_natom','bandgap_energy_ev']
mn = pd.DataFrame(0, index=np.arange(len(nn)), columns=fl)
```

```python
mn['id'] = rf['id']
```

```python
mn['formation_energy_ev_natom'] = 0.4 * nn['formation_energy_ev_natom'] + 0.6 * rf['formation_energy_ev_natom']
mn['bandgap_energy_ev'] = 0.6 * nn['bandgap_energy_ev'] + 0.4 * rf['bandgap_energy_ev']
mn.head()
```

```python
mn.to_csv('ensemble3_rf_nn.csv',index=False)
```
